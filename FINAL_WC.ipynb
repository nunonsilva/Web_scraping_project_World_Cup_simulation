```python
#import necessary libraries
import pandas as pd
import numpy as np
from random import sample
```

```python
# setting the variables for number of simulations, and respective weight of each parameter for the "match simulator" function

N = 5000
player_weight = 0.60
h2h_weight = 0.05
random_weight = 0.35
std = 10 
```

```python
# Grabbing the data obtained through web scraping and the FIFA api

fifa = pd.read_csv("players_20.csv")
confederations = pd.read_csv("organizations.csv")
games = pd.read_csv("game_matrix.csv")
```

```python
# Serbia was missing from the orgs dataset
serb = {"Organization":"UEFA", "Country": "Serbia"}
confederations = confederations.append(serb, ignore_index=True)
```

```python
# Filtering the countries with more than 15 players

viable_countries= fifa["nationality"].value_counts()
viable_countries = viable_countries[viable_countries < 15]
viable_countries= viable_countries.index
viable_countries= list(viable_countries)
```

```python
fifa = fifa[~fifa['nationality'].isin(viable_countries)]
```

```python
# Grouping the best 20 players per country (15-20 if they don't have 20 players) and finding the squad strenght

team_averages = fifa.groupby("nationality")["overall"].nlargest(20).mean(level=0).reset_index(name="overall")
```

```python
team_averages = team_averages.merge(confederations, how='left', left_on='nationality', right_on='Country')
```

```python
team_averages.drop("nationality", axis=1, inplace= True)
```

```python
organization_df = {"Organization":["CONMEBOL", "AFC","CONCACAF","OFC","CAF","UEFA"],"Spots":[4,4,3,0,5,13]}
```

```python
organization_df = pd.DataFrame(organization_df)
```

```python
games.columns = games.columns.str.lstrip()
games.columns = games.columns.str.rstrip()
```

```python
games["Country"]= games["Country"].str.lstrip()
games["Country"]= games["Country"].str.rstrip()
```

```python
games=games.merge(team_averages, how="left", on="Country")
```

```python
games.set_index('Country',drop=False, inplace=True)
```

```python
qualified_teams= pd.DataFrame(games)
```

```python
qatgames= {"overall":60, "Organization": "AFC"}
qualified_teams.fillna(value=qatgames, axis=0, limit=1, inplace=True)
```

```python
qualified_teams
```

```python
games.replace(" NA ",0, inplace=True)
games.replace("#VALUE!",0, inplace=True)
```

```python
# "Match simulator" function used to determine the tournament outcome

def compare_teams(country1, country2): 
  country1_average = float(qualified_teams.loc[country1,"overall"])
  country2_average = float(qualified_teams.loc[country2,"overall"]) 
  country1_h2h = float(games.loc[country1,country2])
  country2_h2h = float(games.loc[country2,country1])

  if country1_h2h is None:
    country1_score = player_weight * country1_average + (1 - player_weight) * np.random.normal(1, 0, std)
    country2_score = player_weight * country2_average + (1 - player_weight) * np.random.normal(1, 0, std)
  else: 
    country1_score = player_weight * country1_average + country1_h2h * h2h_weight + country1_average + random_weight * np.random.normal(0, std, 1)
    country2_score = player_weight * country2_average + country2_h2h * h2h_weight + country2_average + random_weight * np.random.normal(0, std, 1)
  
  if country1_score > country2_score:
    return country1
  else:
    return country2
```

```python
# Setting up a counter dataframe that will count the times that each team makes it through the group stage

wins = {"Country": qualified_teams["Country"], "Wins": np.repeat(0,len(qualified_teams["Country"]))}
wins = pd.DataFrame(wins)
```

```python
# Setting up a counter dataframe that will count the times that each team wins the World Cup

whole_wins = {"Country": qualified_teams["Country"], "Wins": np.repeat(0,len(qualified_teams["Country"]))}
whole_wins = pd.DataFrame(whole_wins)
whole_wins
```

```python
wins.reset_index(drop=True, inplace=True)
```

```python
# Simulating the tournament 5000 times

for i in range(0,N):
  print("Iteration", i, "of", N,".")
  new_final = qualified_teams.sample(frac=1)
  group1 = new_final.iloc[range(0,4)]
  group2 = new_final.iloc[range(4,8)]
  group3 = new_final.iloc[range(8,12)]
  group4 = new_final.iloc[range(12,16)]
  group5 = new_final.iloc[range(16,20)]
  group6 = new_final.iloc[range(20,24)]
  group7 = new_final.iloc[range(24,28)]
  group8 = new_final.iloc[range(28,32)] 
  groups = [group1,group2,group3,group4,group5,group6,group7,group8]
  group_winners= []
  for group in groups:
    random_teams = np.random.choice(range(0,4),2,replace=True)
    group["overall"].iloc[random_teams[0]] = group["overall"].iloc[random_teams[0]]+1
    group["overall"].iloc[random_teams[1]] = group["overall"].iloc[random_teams[1]]+1
    sorted_group = group.sort_values(by="overall", ascending=False)
    wins.loc[wins['Country']==sorted_group["Country"].iloc[0], 'Wins'] += 1
    wins.loc[wins['Country']==sorted_group["Country"].iloc[1], 'Wins'] += 1
    group_winners.append(sorted_group["Country"].iloc[0])
    group_winners.append(sorted_group["Country"].iloc[1])

  round1_winners = []
  a1 = group_winners[0] #A1
  b2 = group_winners[3] #B2
  w = compare_teams(a1, b2)
  round1_winners.append(w)

  a2 = group_winners[1] #A2
  b1 = group_winners[2] #B1
  w = compare_teams(a2, b1)
  round1_winners.append(w)

  c1 = group_winners[4] #C1
  d2 = group_winners[7] #D2
  w = compare_teams(c1, d2)
  round1_winners.append(w)
    
  c2 = group_winners[5] #C2
  d1 = group_winners[6] #D1
  w = compare_teams(c2, d1)
  round1_winners.append(w)

  e1 = group_winners[8] #E1
  f2 = group_winners[11] #F2
  w = compare_teams(e1, f2)
  round1_winners.append(w)

  e2 = group_winners[9] #E2
  f1 = group_winners[10] #F1
  w = compare_teams(e2, f1)
  round1_winners.append(w)

  g1 = group_winners[12] #G1
  h2 = group_winners[15] #H2
  w = compare_teams(g1, h2)
  round1_winners.append(w)

  g2 = group_winners[13] #G2
  h1 = group_winners[14] #H1
  w = compare_teams(g2, h1)
  round1_winners.append(w) 

  round2_winners = []

  a1b2 = round1_winners[0]
  c1d2 = round1_winners[2]
  w = compare_teams(a1b2, c1d2)
  round2_winners.append(w)

  a2b1 = round1_winners[1]
  c2d1 = round1_winners[3]
  w = compare_teams(a2b1, c2d1)
  round2_winners.append(w)

  e1f2 = round1_winners[4]
  g1h2 = round1_winners[6]
  w = compare_teams(e1f2, g1h2)
  round2_winners.append(w)

  e2f1 = round1_winners[5]
  g2h1 = round1_winners[7]
  w = compare_teams(e2f1, g2h1)
  round2_winners.append(w) 

  round3_winners = []
  c1 = round2_winners[0]
  c2 = round2_winners[2]
  w = compare_teams(c1, c2)
  round3_winners.append(w)
  
  c3 = round2_winners[1]
  c4 = round2_winners[3]
  w = compare_teams(c3, c4)
  round3_winners.append(w)

  c1 = round3_winners[0]
  c2 = round3_winners[1]
  wm = compare_teams(c1, c2)
  whole_wins.loc[whole_wins['Country']==wm, 'Wins'] += 1
```
